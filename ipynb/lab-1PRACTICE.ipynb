Write a program that keeps asking the user for a number until they enter a valid integer.
Then print: 
 You entered: X
Square: X²

```python
while True:
 a= input("input a valid number: ")
 if a.isdigit():
    
    Square = int(a)**2
    print ("You entered :"+a+"\nsquare:"+str(Square) )

    break
 else: print ("Invalid input input again")

```

Ask user birth year and print:
Child (<13);
Teen (13–19);
Adult (20–59);
Senior (60+);
👉 Must validate numeric input.

```python
year_s= input ("input Your Age")
if year_s.isdigit():
    year= int (year_s)
    if year <13:
        print ("child")
    if year >=13 and year <=20:
        print ("teen")
    if year >20 and year <= 60:
        print ("Adult")
    if year>60:
        print ("senior")
else: print ("Invalid input input again")
```
