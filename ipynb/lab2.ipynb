Agent: Perception ---> action
[A,clean] ----> Right
[A,darty] ----> Suck
[B,clean] ----> Left
[B,darty] ----> Suck

```python
def simple_reflex_agent(percept):
    print ("preception Recived:"+str(percept) ) # percept ['location' clean or not]
    location = percept[0]
    status = percept[1]
    if status == 'Dirty':
        action = "suck"
    elif location == 'A':
        action = "Right"
    elif location == 'B':
        action = "Left"
    return action
```

```python
import random
location = random.choice(['A', 'B'])
status = random.choice(['Clean', 'Dirty'])
while True:
    action = simple_reflex_agent([location, status])
    print ("Action Performed:"+action) 
    cmd = input("get pyerception (yes/no): ")
    if cmd == "no" or cmd != "yes":
        break
    if action == "Right":
        location = 'B'
        condition = random.choice(['Clean', 'Dirty'])
    elif action == "Left":
        location = 'A'
        condition = random.choice(['Clean', 'Dirty'])
    else:
        condition = 'Clean'
```
