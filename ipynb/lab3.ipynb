```python
model ={'A':'unknown','B':'unknown'}
world_state= 'bad'
action = 'unknown'
def update_state(action,percept,model):
    location= percept[0]
    status= percept[1]
    model[location]= status
    print("perceptiom: "+str(percept))
    print("model before action: "+str(model))
    global world_state
    #UPDATE THE MODEL BASE AVTION
    if action == 'suck':
        model[location]= 'clean'
    #model checking to update the world state
    if model ['A']== 'clean' and model['B']=='clean':
        world_state= 'good'
    else:
        world_state= 'bad'
    return world_state
def model_based_reflex_agent(percept):
    location= percept[0]
    status= percept[1]
    global world_state, action, model
    if world_state == 'good':
        action='pause'
        return action
    elif status == 'dirty':
        action='suck'
       
    elif location == 'A':
        action='right'
    elif location == 'B':
        action='left'
    world_state= update_state(action,percept,model)
    print("action perfomed:"+action)
    print('model after action: '+str(model))
    print('state: '+world_state)
    return action



```

```python
import random
location = random.choice(['A', 'B'])
condition = 'dirty'
while True:
    print('*********')
    action = model_based_reflex_agent((location,condition))
   
    if action == "right":
        location = 'B'
        condition = 'dirty'
    elif action == "left":
        location = 'A'
        condition = 'dirty'
    elif action == "suck":
        condition = 'clean'
    elif action == "pause":
        cmd = input("STOPED , Do you want to continue? (yes/no): ")
        if (cmd== 'no' or cmd !='yes'):
            break
        location = random.choice(['A', 'B'])
        condition = 'dirty'

        model ={'A':'unknown','B':'unknown'}
        world_state= 'bad'
        action = 'unknown'
```
