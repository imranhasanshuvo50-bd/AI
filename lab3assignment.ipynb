```python
robot_memory = {'room1': 'unknown', 'room2': 'unknown', 'room3': 'unknown' }
home_environment = 'dirty'
next_action = 'unknown'

def update_memory(next_action,location_state ,robot_memory):
     location = location_state[0]
     state = location_state[1]
     robot_memory[location] = state
     print("perceptiom: "+str(location)+ ","+str(state))
     print("model before action: "+str(robot_memory))
     if next_action== 'suck':
        robot_memory[location] = 'clean'
def models_base_reflex_agent(location_with_state):
    global robot_memory, home_environment, next_action
    location= location_with_state[0]
    state = location_with_state[1]
    if home_environment == 'clean':
        next_action = 'stop'
        return next_action
        
    elif state == "dirty":
         next_action = 'suck'
         
    elif location == 'room1'and robot_memory['room2'] == 'unknown':
         next_action = "right"
    elif location == 'room3' and robot_memory['room2'] == 'unknown':
         next_action = "left"
    elif location == 'room2' and robot_memory['room3'] == 'unknown':
         next_action = "right"
    elif location == 'room2' and robot_memory['room1'] == 'unknown':
         next_action = "left"
    elif location == 'room3' and robot_memory['room1'] == 'unknown':
         next_action = "right"

    elif robot_memory['room1'] == 'clean' and robot_memory['room2'] == 'clean' and robot_memory['room3'] == 'clean':
         home_environment = 'clean'
         #next_action = 'stop'
     
    update_memory(next_action,(location, state),robot_memory)
    print("action perfomed:"+next_action)
    print('model after action: '+str(robot_memory))
    print('state: '+home_environment)
    return next_action

```

```python
import random
start = random.choice(['room1','room3'])
location = start
condition = 'dirty'
while True:
    print('*********')
    action = models_base_reflex_agent((location, condition))
   
    if action == "right" and location == 'room1':
        location = 'room2'
        condition = 'dirty'
    elif action == "left" and location == 'room3':
        location = 'room2'
        condition = 'dirty'
    elif action == 'right' and robot_memory['room3'] == 'unknown':
          location = 'room3'
          condition = 'dirty'
    elif action == 'left' and robot_memory['room1'] == 'unknown':
          location = 'room1'
          condition = 'dirty'
     
    elif action == "suck" :
      condition = 'clean'
    elif action == "stop":
        cmd = input("STOPED , Do you want to continue? (yes/no): ")
        if (cmd== 'no' or cmd !='yes'):
            break
        start = random.choice(['room1', 'room3'])
        location = start
        condition = 'dirty'

        robot_memory = {'room1': 'unknown', 'room2': 'unknown', 'room3': 'unknown' }
        home_environment = 'dirty'
        next_action = 'unknown'
```
